```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os

path = 'D:\One_한양대학교\private object minsu\coding\data\samsung_2024\\food_preliminary\\ver3'
sub_list = os.listdir(path)

self_rating = np.zeros((len(sub_list), 5), int)
for i, sub in enumerate(sub_list):
    dat_list = os.listdir(path + '/' + sub)
    for dat in dat_list:
        if dat[:4] == 'Food':
            a = pd.read_csv(path + '/' + sub + '/' + dat).to_numpy()[0]
            for j in range(1,6):
                self_rating[i, j-1] += sum(a==j)

avg_self_rating = self_rating/len(sub_list)

def plot_self_rating(self_rating):
    plt.figure(figsize=(8,4))
    plt.bar(np.arange(0,5,1), self_rating, 0.5, color=(0.5, 0.5, 0.5, 0.6))
    plt.xticks(np.arange(0,5,1), np.arange(1,6,1), fontweight='bold', fontsize='x-large', fontfamily='Malgun Gothic')
    # plt.yticks(np.arange(0,15,1), fontweight='semibold',fontsize='x-large', fontfamily='Malgun Gothic')
    plt.ylabel("Averaged Selected Points", fontsize='xx-large', fontweight='bold', fontfamily='Malgun Gothic')
    plt.xlabel("Preference Points", fontsize='xx-large', fontweight='bold', fontfamily='Malgun Gothic')
    for i, value in enumerate(self_rating):
        plt.text(i, value - max(self_rating)/10, str(value), ha='center', fontsize=15)

for i in range(5):
    plot_self_rating(self_rating[i])
plot_self_rating(np.mean(self_rating,0))
```
